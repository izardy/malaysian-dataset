```python
from bs4 import BeautifulSoup
from glob import glob
```

```python
files = glob('*.html')
```

```python
videos = []
for f in files:
    with open(f) as fopen:
        htm = fopen.read()

    bs = BeautifulSoup(htm)
    videos.extend(bs.find_all('a'))
```

```python
videos = [a_.get('href') for a_ in videos if '/watch?v=' in a_.get('href', '')]
```
