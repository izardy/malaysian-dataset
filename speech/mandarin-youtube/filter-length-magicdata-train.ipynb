```python
from glob import glob
from tqdm import tqdm
import soundfile as sf
import mp
```

```python
mandarin = glob('train/*/*.wav')
len(mandarin)
```

```python
def loop(files):
    files, _ = files
    filtered = []
    for f in tqdm(files):
        y, sr = sf.read(f)
        if len(y) / sr >= 5:
            filtered.append(f)
    return filtered
```

```python
results = mp.multiprocessing(mandarin, loop, cores = 50)
```
