```python
from glob import glob
import os

files = glob('*/*.wav')
len(files)
```

```python
import pandas as pd

df = pd.read_csv('class_labels_indices.csv')
mapping = {}
labels = {}
for i in range(len(df)):
    mapping[df['mid'].iloc[i]] = df['index'].iloc[i]
    labels[df['index'].iloc[i]] = df['display_name'].iloc[i]
```

```python
data = []

files = ['balanced_train_segments.csv', 'eval_segments.csv']

for f in files:
    f_only = f.replace('.csv', '')
    with open(f) as fopen:
        for l in fopen:
            l = l.strip()
            splitted = l.split(', ')
            audio_filename = os.path.join(f_only, f'Y{splitted[0]}.wav')
            if not os.path.exists(audio_filename):
                continue
            data.append({
                'audio_filename': audio_filename,
                'start': float(splitted[1]),
                'end': float(splitted[2]),
                'labels': [int(mapping[i]) for i in splitted[3][1:-1].split(',')],
            })
len(data)
```

```python
import json

with open('populate.json', 'w') as fopen:
    json.dump(data, fopen)
```

```python
import soundfile as sf

for d in data:
    y, sr = sf.read(d['audio_filename'])
    break
```
